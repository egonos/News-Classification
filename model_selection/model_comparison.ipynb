```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
```

```python
evals = pd.read_csv("model_logs.csv",index_col=0)

review_evals = evals[evals["Dataset"] == "REVIEWS"]
imdb_evals = evals[evals["Dataset"] == "IMDB"]
amazon_evals = evals[evals["Dataset"] == "AMAZON"]

palette = {'distilbert-base-uncased-finetuned-sst-2-english': "#b0b0b0",
       'AirrStorm/DistilBERT-SST2-Yelp': "#7b7c85",
       'siebert/sentiment-roberta-large-english': "#003f7f"}
```

```python
plt.title("Model Performance - Balanced Dataset")
sns.scatterplot(x = "Inference Time", y = "ROC-AUC Score",data = imdb_evals, hue="Model Name",s = 70,palette=palette)
plt.show()
```

```python
plt.title("Model Performance - Positive Imbalanced Dataset")
sns.scatterplot(x = "Inference Time", y = "ROC-AUC Score",data = review_evals, hue="Model Name",s = 70,palette=palette)
plt.show()
```

```python
plt.title("Model Performance - Negative Imbalanced Dataset")
sns.scatterplot(x = "Inference Time", y = "ROC-AUC Score",data = amazon_evals, hue="Model Name",s = 70,palette=palette)
plt.show()
```
